```python
import uuid
import json
from readingdb.annotator import Annotator
from readingdb.constants import AnnotatorKeys, Database, REGION_NAME
from botocore.config import Config
from readingdb.endpoints import BUCKET, DYNAMO_ENDPOINT
from readingdb.api import API
```

```python
api = API(
    DYNAMO_ENDPOINT,
    size_limit=999999,
    bucket=BUCKET,
    tmp_bucket=BUCKET,
    config=Config(
        region_name=REGION_NAME
    )
)
```

```python
uid = "06eca9ee-fa0c-42b1-b477-42e433aba6af"
```

```python
routes = api.routes_for_user(uid)
```

```python
len(routes)
```

```python
pth = '/home/lewington/Desktop/vbackup/'
```

```python
for r in routes:
    rid = r['RouteID']
    readings = api.all_route_readings(rid, uid)
    
    with open(pth + f'/{rid}.json', 'w') as f:
        print(readings[:20])
        json.dump(readings, f)
    print("finished", rid)
```
